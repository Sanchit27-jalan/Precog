```python
import gensim.downloader as api
import numpy as np
from scipy.spatial.distance import cosine
from scipy.stats import spearmanr,pearsonr
from gensim.models import KeyedVectors
```

```python
# Load pre-trained Word2Vec model
model = KeyedVectors.load_word2vec_format('./GoogleNews-vectors-negative300.bin', binary=True)

# Normalize the vectors
model.init_sims(replace=True)
```

```python
def word_similarity(word1, word2):
    try:
        embedding1 = model[word1]
        embedding2 = model[word2]
        similarity = 1 - cosine(embedding1, embedding2)
        return similarity
    except KeyError:
        return None  # Return None if any of the words are not in the vocabulary
```

```python
word_pairs = []
human_scores = []

with open('my.txt', 'r') as file:
    for line in file:
        word1, word2, score = line.split()
        word_pairs.append((word1, word2))
        human_scores.append(float(score))

print(word_pairs)
print(human_scores)
```

```python
predicted_scores = [word_similarity(pair[0], pair[1]) for pair in word_pairs]

predicted_scores = [score for score in predicted_scores if score is not None]
```

```python
human_scores = human_scores[:len(predicted_scores)]

predicted = np.array(predicted_scores)
actual = np.array(human_scores)

mse = np.mean((predicted - actual) ** 2)
mae = np.mean(np.abs(predicted - actual))
rmse = np.sqrt(np.mean((predicted - actual) ** 2))
pearson_corr, _ = pearsonr(predicted, actual)
spearman_corr, _ = spearmanr(predicted, actual)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("Root Mean Squared Error:", rmse)
print("Pearson Correlation Coefficient:", pearson_corr)
print("Spearman Rank Correlation Coefficient:", spearman_corr)
```
